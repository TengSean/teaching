```python
from bs4 import BeautifulSoup

from urllib.request import urlopen, urljoin
import re, os, requests
import numpy as np
import pandas as pd


response = urlopen("https://ww2.money-link.com.tw/TWStock/StockBasic.aspx?TWMId=Basic_drt03&SymId=4994")
html = response.read().decode('utf-8')
print(html)
```

```python
bs4 = BeautifulSoup(html)
# print(bs4)
tr = bs4.find_all('tr')
# print(tr)
colName1 = ['設定日期', '身份', '姓名', '目前持股(張)', '設質(張)', '解質(張)', '張數', '佔發行(%)', '佔持股(%)', '質權人']
colName2 = ['設定日期', '身份', '姓名', '目前持股(張)', '設質(張)', '解質(張)', '張數', '佔發行(%)', '佔持股(%)', '質權人']
dataNP = -1
for td in tr:
    
    if len(td) == 21:
        a = np.array((td.text).strip().split('\n'))[None,:]
        if dataNP == -1:
            dataNP = a
        else:
            dataNP = np.vstack((dataNP, a))
df = pd.DataFrame(dataNP)
df1 = df[:35]
df1.columns = colName1
df2 = df[35:]
# df1.columns = colName2
display(df1)
display(df2)
# # print(xls)
```
